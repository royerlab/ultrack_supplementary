```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

#           FN          FP          add      del        semantic
labels = ['FN_node', 'FP_node', 'FN_edge', 'FP_edge','semantic']
ultrack_means = [5,250,16,7,1]
trackmate_means = [40,508,130,28,15]

labels = labels[::-1]
ultrack_means = ultrack_means[::-1]
trackmate_means = trackmate_means[::-1]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(111)
rects1 = ax.barh(x - width/2, trackmate_means, width, label='TrackMate')
rects2 = ax.barh(x + width/2, ultrack_means, width, label='Ultrack')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Tracking mistakes')
# ax.set_title('Scores by group and gender')
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()
# ax.set_xscale('log')
plt.xlim((0,550))

# plt.ylim((0,1400))
# plt.grid()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


# autolabel(rects1)
# autolabel(rects2)

fig.tight_layout()



FIG_DIR = Path('./500_frames/')
filename = "neuromast_TRA.pdf"
plt.savefig(FIG_DIR / filename)

plt.show()
```
